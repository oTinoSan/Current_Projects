```python
import pandas as pd
import sqlite3
import os

# Define the file and path
fileName1 = '20090919071137.csv'
path = './Geolife_Trajectories_1.3/Data/120/Trajectory/'
FileIdentification1 = os.path.join(path, fileName1)

# Connect to SQLite database (or create it if it doesn't exist)
con = sqlite3.connect('geolife.db')  # connection to database established
cur = con.cursor()

# Read CSV file and directly load it into the SQLite database
# The table name is set as 'Points_User120_Table1'
# if_exists='replace' will drop the table if it exists and create a new one.
# Adjust if_exists as needed: 'append' to add to an existing table, 'fail' to do nothing if the table exists.
# index=False ensures that pandas does not insert its own index into the SQLite table.
pd.read_csv(FileIdentification1).to_sql('Points_User120_Table1', con, if_exists='replace', index=False)  # Corrected 'conn' to 'con'

# Example query to print all data from Points_User120_Table1
print('Query 1')
print('List all data in Points_User120_Table1')
print("Query 1 Results")
cur.execute('SELECT * FROM Points_User120_Table1 LIMIT 10')
for record in cur.fetchall():  # Use fetchall() to retrieve all records
    print(record)
print('')



```

```python
import pandas as pd
import sqlite3
import os

fileName1 = '20090919071137.csv' 
path = './Geolife_Trajectories_1.3/Data/120/Trajectory/' 
FileIdentification1 = os.path.join(path, fileName1)

# Read the CSV file into a DataFrame
Points_User120_File1 = pd.read_csv(FileIdentification1)
Points_User120_File1 = Points_User120_File1.drop(columns=['Order', 'Unused'])

conn = sqlite3.connect('geolife.db')

conn.execute("""
CREATE TABLE IF NOT EXISTS User120_Trajectory (
    Latitude REAL,
    Longitude REAL,
    Unused INTEGER,
    Altitude INTEGER,
    Days TEXT,
    Date TEXT,
    Time TEXT
)
""")

cur = conn.cursor()

Points_User120_File1.to_sql('User120_Trajectory', conn, if_exists='append', index=False)

conn.close()
```

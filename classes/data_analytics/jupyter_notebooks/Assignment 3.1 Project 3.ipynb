# Task 2:  Concatenating  Two DataFrames - Union Operation

### General Information:
* Concatenating means combining two DataFrames into a single DataFrame. 
* In general there are cases of simple and complex concatenation. 
* Simple concatenation is when there is a smooth transition between the last item in the first file and the first item of the second file. 
* Complex concatenation is when a special technique needs to be applied.

There are several techniques to accomplish complex concatentation:
1. Use additional column with the file name since the file name uniquly identifes sub-directory.
2. Use additional row to indicate the sentinel
3. Use double indexing to provide a global index and local within trajectory index.

### Specific Information:
* The last item in the file  designates the end of sub-trajectory and the beginning of the new file can  represent the begining of completely new sub-directory with the space and time  jump hard to interpret.
* This transition can polute any statistics about time and speed of travel.  In the code below we have used the last method for several reasons: 
    * we could use regular concatenate plus minor modifications
    * our files are heavily indexed anyway since our processing requires indexing for speed
    * for Big Data we can save space without additional attribute
    * indexes are more understandable to humans for visual inspections
* In our case  combining Points_User121_File1 and Points_User121_File1 will result in a single DataFrame, Points_User121.
* It can be accomplished by the following code:

```python
# Load data for a single trajectory for a specific user and display the first rows 
Points_User121 = pd.concat([Points_User121_File1, Points_User121_File2], ignore_index=True)
Points_User121
```

### General Information:
* The code for Identifying  file names in the given Directory can be extended for Creating a User define Function for Reading and Concatenating multiple files in the given Directory 

### Specific Information:
* In our specific case, we will define a function that will read in the csv files for a given directory
* It will also add these files into a final DataFrame
* At the end, it will return that DataFrame to us, allowing us to preserve it in a new variable name.

```python
def readDataFromDirectory(userId):
    dataPath = './Geolife_Trajectories_1.3/data/' + userId + '/Trajectory'
    print('dataPath:', dataPath)
    print()
    table = pd.DataFrame() #create an empty table
    for file in os.listdir(dataPath):
        if file.lower().endswith(".csv"): # only work with csv files, because other file types would crash program
            print("csv file:", file)
            FileIdentification = dataPath +'/'+ file
            print('FileIdentification:',FileIdentification)
            tempTable = pd.read_csv(FileIdentification) # fill temporary table with data from new file
            table = pd.concat([table, tempTable], ignore_index=True) # concatenate final with temp, in a loop.
            print("Added", file, "to table")
            print()
    print()
    return(table)
```

```python
# function call
userId='120'
Points_User120 = readDataFromDirectory(userId) 
Points_User120
```

# Task 4: Reading files from Multiple Directories and adding appropriate subdirectory identifiers

### General Information:

* The Path provides a directory of files to add to a table to be more efficient by using the os library.

### Specific Information:

* Use the os method listdir(path)- to get the list of directories of the path
* This will be the Users
* From Task 3 update the code using the proper os methods.

```python
def readDataFromDirectories(path='./Geolife_Trajectories_1.3/Data/'):
    table = pd.DataFrame()
    for dir in os.listdir(path):
        print("Adding user", dir, "to table")
        tempTable = readDataFromDirectory(dir)
        tempTable['User']= str(dir) # to add the username for this directory
        table = pd.concat([table, tempTable], ignore_index=True)
    print()
    return table
```

```python
Points_All3 = readDataFromDirectories()
Points_All3
```

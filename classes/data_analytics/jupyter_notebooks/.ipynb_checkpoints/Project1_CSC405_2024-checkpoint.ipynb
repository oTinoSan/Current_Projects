```python
def selectRowsCondition(table, columnName, operator, value):
    
    operatorList = ['>', '<', '='] 
    if operator not in operatorList:
        print("Please use one of the following operators. ['>', '<', '='] ")
        newTable = table

    else:
        if operator == '>':
            condition = table[columnName] > value
        elif operator == '<':
            condition = table[columnName] < value
        else:
            condition = table[columnName] = value
        newTable = table[condition]
    return (newTable)
```

```python
PointsTable_User120_File1_Limited = selectRowsCondition(PointsTable_User120_File1, 'Latitude', '>', 42.5)

PointsTable_User120_File1_Limited
```

9.1 Bar Chart showing the total number of recordings when the Users 120 was close and when the Users 120 was far from the airport.

9.2 Pie Chart showing the total number of recordings when the Users 120 was close and when the Users 120 was far from the airport.






```python
import matplotlib.pyplot as plt

initialFrame = readDataFromFile('20090919071137.csv','120')
print(initialFrame)

table2 = selectRowsCondition(table1, 'Latitude', '>', 42.0)
table3 = selectRowsCondition(table2, 'Latitude', '<', 42.2)
table4 = selectRowsCondition(table3, 'Longitude', '>', 117.4) 
closeFrame = selectRowsCondition(table4, 'Longitude', '<', 117.6)
print(closeFrame)

mergedFrames = pd.merge(initialFrame, closeFrame, how='left', indicator=True)
filterFrames = mergedFrames[mergedFrames['_merge'] == 'left_only']
farFrame = filterFrames.drop(columns=['_merge'])
print(farFrame)

# bar chart

# Getting the number of rows for each DataFrame
total_close = closeFrame.shape[0]
total_far = farFrame.shape[0]

# Plotting
labels = ['User 120 Close', 'User 120 Far']
values = [total_close, total_far]

plt.bar(labels, values, color=['blue', 'green'])

# Adding labels and title
plt.xlabel('Close and Far')
plt.ylabel('Number of Occurances')
plt.title('Comparison of Close and Far Recordings')

# Display the plot
plt.show()
```
